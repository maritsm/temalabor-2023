```python
coh_df = getCohDf()
```

```python
coh_df
```

```python
coh_df.index = pd.to_datetime(coh_df.index)
monthly = coh_df.groupby(pd.Grouper(freq = "M"))
```

```python
aggr = monthly.agg("mean")
```

```python
aggr
```

```python
for i in aggr.columns:
    aggr[i] = aggr[i].ffill()
aggr.plot(figsize=(20,20), linestyle='-', marker='.')
```
