```python
proc = cosp.groupby("procedure_interinst_id")
```

Nyilván minden procedure-höz tartozik több `date` is, mert miért lenne egyszerű? :)
Fogom és veszem az összesnek a minimumát, alias kezdődátumát.
Ezeket kiírom egy új csv-be.

```python
for i, x in proc:
    print(i, min(x["date"].unique()))
```

```python
procedure_dates = []
for i, x in proc:
    procedure_dates.append([i, min(x["date"].unique())])

procdate_df = pd.DataFrame(procedure_dates)
```
